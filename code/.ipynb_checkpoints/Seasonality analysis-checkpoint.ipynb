```python
from functools import partial
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as st
from sklearn.utils import resample
import pandas as pd
from scipy.optimize import curve_fit
import pickle
import seaborn as sns
import calendar
from dateutil.relativedelta import relativedelta
```

```python
def normalize_by_fitting(data, t, sigma=None, bounds=(-np.inf, np.inf), plot=False, exp=True):
    '''
    normalizes the data by canceling the exponential component of it
    
    @params:
        data - [#time X #samples]
        sigma - [#time X #samples], should contain values of standard deviations of errors in data.
                In this case, the optimized function is ``chisq = sum((r / sigma) ** 2)``
        bounds - 2-tuple of array_like, optional
                 Lower and upper bounds on parameters. Defaults to no bounds.
                 Each element of the tuple must be either an array with the length equal
                 to the number of parameters, or a scalar (in which case the bound is
                 taken to be the same for all parameters.) Use ``np.inf`` with an
                 appropriate sign to disable bounds on all or some parameters.
        
    @return:
        normalized data - data.shape
        fitted params - [#samples X #params]
        params_std
    '''
    
    _sigma = sigma
    linear = lambda t, A, alpha: np.log(A) - alpha*t

    fitted_params = np.zeros((data.shape[1], 2))
    params_std = np.zeros_like(fitted_params)
    data_normalized = np.zeros_like(data)
    for i in range(data.shape[1]):
        y = np.log(data[:, i])
        if sigma is not None:
            _sigma = sigma[:, i]
        popt, pcov = curve_fit(linear, t[np.isfinite(y)], y[np.isfinite(y)], sigma=_sigma, bounds=bounds)
        y[~np.isfinite(y)] = linear(t, *popt)[~np.isfinite(y)]
        data_normalized[:, i] = y - linear(t, *popt)
        if exp:
            data_normalized[:, i] = np.exp(data_normalized[:, i])
        fitted_params[i] = popt
        params_std[i] = np.sqrt(np.diag(pcov))
            
    if plot:
        plt.figure(figsize=(4*2,3*2))
        plt.plot(np.arange(1, 7), data_normalized)
        plt.xlabel('hair segment', size=14)
        plt.ylabel('normalized cortisol', size=14)
        plt.show()
        
        f, axs = plt.subplots(2, 3, figsize=(4*3, 3*2), sharex=True)
        axs_f = axs.flatten()
        for i, ax in enumerate(axs_f):
            sns.distplot(data_normalized[i], ax=ax)
        plt.show()
        
        N_boot = 1000
        segment_means = np.zeros((N_boot, data_normalized.shape[0]))
        segment_stds = np.zeros((N_boot, data_normalized.shape[0]))
        for i in range(N_boot):
            bootstrap_data = data_normalized[:, resample(np.arange(data_normalized.shape[1]))]
            segment_means[i] = np.mean(bootstrap_data, axis=1)
            segment_stds[i] = np.std(bootstrap_data, axis=1)

        plt.figure()
        plt.errorbar(np.arange(1, 7), np.mean(segment_means, axis=0), np.std(segment_means, axis=0), fmt='.-')
        plt.xlabel('hair segment', size=14)
        plt.ylabel('mean', size=14)
        plt.show()

        plt.figure()
        plt.errorbar(np.arange(1, 7), np.mean(segment_stds, axis=0), np.std(segment_stds, axis=0), fmt='.-')
        plt.xlabel('hair segment', size=14)
        plt.ylabel('std', size=14)
        plt.show()
        
        f, ax = plt.subplots(2, 2, figsize=(4*2, 3*2))
        ax[0, 0].hist(fitted_params[:, 0], bins='auto')#, density=True)
        ax[0, 0].set_xlabel('A')
        ax[0, 1].scatter(fitted_params[:, 0], params_std[:, 0]/np.abs(fitted_params[:, 0]))
        ax[0, 1].set_xlabel('A')
        ax[0, 1].set_ylabel('relative error ($\sigma_A/A$)')
        
        
        ax[1, 0].hist(fitted_params[:, 1], bins='auto')#, density=True)
        ax[1, 0].set_xlabel('$\\alpha$')
        ax[1, 1].scatter(fitted_params[:, 1], params_std[:, 1]/np.abs(fitted_params[:, 1]))
        ax[1, 1].set_xlabel('$\\alpha$')
        ax[1, 1].set_ylabel('relative error ($\sigma_\\alpha/\\alpha$)')
        ax[1, 1].set_xlim((min(fitted_params[:, 1]), max(fitted_params[:, 1])))
        ax[1, 1].set_yscale('log')
        
        plt.tight_layout()
        plt.show()
        print('Mean personal decaying constant = {:.3f}'.format(np.mean(fitted_params[:, 1])))
        print('STD of personal decaying constants = {:.3f}'.format(np.std(fitted_params[:, 1])))
        print('SEM = {:.5f}'.format(st.sem(fitted_params[:, 1])))

    return data_normalized, fitted_params, params_std
```

```python
hair_cortisol_data_path = '../data/hair cortisol - 140120.xlsx'
data_to_analyze = pd.read_excel(hair_cortisol_data_path, sheet_name='proccessed', index_col=0, nrows=10)
data_to_analyze = data_to_analyze.T.reset_index()
data_to_analyze.rename(columns={'index': 'sample ID'}, inplace=True)
data_to_analyze = pd.melt(data_to_analyze, id_vars=['sample ID', 'subject ID', 'type', 'cut_date', 'age'], var_name='segment')
data_to_analyze['value'] = pd.to_numeric(data_to_analyze['value'])
data_to_analyze['segment'] = pd.to_numeric(data_to_analyze['segment'])
data_to_analyze['sample ID'] = data_to_analyze['sample ID'].astype(str)
data_to_analyze['age'] = data_to_analyze['age'].fillna(0)
data_to_analyze
```

```python
df = data_to_analyze.copy()
df['cut_date'] = pd.to_datetime(df['cut_date'], format='%d.%m.%y')
# 0.34cm follicle-scalp, 0.8cm scalp-scissors -> follicle-scissors=1.14cm=34days. I am taking the middle point of the
# segment as the refernce date (therefore adding 30 days more)
df['segment_date'] = df['cut_date'] - pd.Timedelta(64, 'day') - pd.Timedelta(60, 'day') * df['segment']
df['segment_month'] = df['segment_date'].dt.month
df['measure_month'] = df['cut_date'].dt.month
df
```

```python
df_organized = df.drop(index=df.index[df['type'].str.contains('r')]).pivot(
    values=['value', 'segment_month', 'segment_date'],
    index='sample ID',
    columns='segment',
).reset_index()
print(df_organized.shape)
df_organized
```

```python
data_mat = df_organized['value'].to_numpy().T.astype(float)
months_mat = df_organized['segment_month'].to_numpy().T.astype(int)
subjects = df_organized['sample ID'].to_numpy()
```

```python
t = np.arange(6) * 60 + 30
data_norm, fitted_params, _ = normalize_by_fitting(data_mat, t, bounds=([0, 0.0001], np.inf), plot=True)
df_organized['A'] = fitted_params[:, 0]
df_organized['alpha'] = fitted_params[:, 1]
```

```python
def sort_by_calendaric_month(data, months):
    '''
    data - [#segments X #participants] normalized hair cortisol
    months - [#segments X #participants] the month of the respective data point in data
    '''
    
    N_participants = data.shape[1]
    sorted_data = np.full((12, N_participants), np.NaN)
    zeroed_months = months - 1
    for i in range(N_participants):
        sorted_data[np.sort(zeroed_months[:, i]), i] = data[np.argsort(zeroed_months[:, i]), i]
    
    return sorted_data

        
def fit_cosinor(data):
    '''
    data - [12 months X #participants]
    returns: mean, amplitude, phase
    '''
    
    mean_per_month = np.nanmean(data, axis=1)
    
    cosinor = lambda t, a, phi, f: a*np.cos(2*np.pi*f*(t + phi))
    cosinor_1year = partial(cosinor, f=1/12)
    
    t = np.arange(12)
    popt, pcov = curve_fit(cosinor_1year, t, mean_per_month - np.mean(mean_per_month), bounds=([0, -np.inf], [np.inf, np.inf]))
    
    return np.mean(mean_per_month), popt[0], popt[1]
```

```python
data_by_month = sort_by_calendaric_month(data_norm, months_mat)
data_by_month_mean, data_by_month_amp, data_by_month_phase = fit_cosinor(data_by_month)
print(data_by_month_mean, data_by_month_amp, data_by_month_phase)
```

```python
# bootstrap participants
N_boots = 10000

boot_mean = np.full(N_boots, np.NaN)
boot_amp = np.full(N_boots, np.NaN)
boot_phase = np.full(N_boots, np.NaN)

for i in range(N_boots):
    idx = resample(np.arange(data_norm.shape[1]))
    boot_data_by_month = sort_by_calendaric_month(data_norm[:, idx], months_mat[:, idx])
    boot_mean[i], boot_amp[i], boot_phase[i] = fit_cosinor(boot_data_by_month)
    
boot_phase[boot_phase > 0] = 12 - boot_phase[boot_phase > 0]
boot_phase[boot_phase < 0] = -boot_phase[boot_phase < 0]
```

### Correcting for seasonality

```python
t = np.linspace(0, 12, 100)
plt.plot(t, cosinor_1year(t, a_data, phi_data) + np.mean(monthly_cortisol))

plt.figure()
plt.plot(t, 1 + cosinor_1year(t, a_data, phi_data) / np.mean(monthly_cortisol))
```

```python
def interpolate_season_factor(date, func, mean_func):

    N_days_in_month_cur = calendar.monthrange(date.year, date.month)[1]
    mid_month_cur = int(N_days_in_month_cur / 2)
    if date.day < mid_month_cur:
        prev_month = date - relativedelta(months=1)
        N_days_in_month_prev = calendar.monthrange(prev_month.year, prev_month.month)[1]
        mid_month_prev = int(N_days_in_month_prev / 2)
        x = (mid_month_prev + date.day) / (mid_month_cur + mid_month_prev) + prev_month.month - 1 # -1 because time is [0, 11]
    else:
        next_month = date + relativedelta(months=1)
        N_days_in_month_next = calendar.monthrange(next_month.year, next_month.month)[1]
        mid_month_next = int(N_days_in_month_next / 2)
        x = (date.day - mid_month_cur) / (mid_month_cur + mid_month_next) + date.month - 1

    return 1 + func(x) / mean_func
```

```python
df_corrected = df_organized.copy()

func = partial(
    interpolate_season_factor,
    func=partial(cosinor_1year, a=a_data, phi=phi_data),
    mean_func=np.mean(monthly_cortisol)
)

for i in np.arange(6):
    df_corrected[('value_norm', i)] = data_norm[i]
    df_corrected[('season_factor', i)] = df_corrected[('segment_date', i)].apply(func)
    df_corrected[('value_season_corrected', i)] = df_corrected[('value_norm', i)] / df_corrected[('season_factor', i)]
```
